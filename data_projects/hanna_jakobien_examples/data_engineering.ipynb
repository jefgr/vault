```python
import pandas as pd
```

```python
df = pd.read_csv('zara.csv', delimiter=';')
```

```python
df.head()
```

```python
df.nunique().head(20)
```

```python
# Drop useless columns
df_cleaned = df.drop(columns=['Product Category', 'brand', 'sku', 'currency', 'description', 'url', 'scraped_at'])
# binary encoding
df_cleaned.replace({'section': {'MAN': False, 'WOMAN': True}}, inplace=True)
df_cleaned['section'] = df_cleaned['section'].astype(bool)

df_cleaned.replace({'Seasonal': {'No': False, 'Yes': True}}, inplace=True)
df_cleaned['Seasonal'] = df_cleaned['Seasonal'].astype(bool)

df_cleaned.replace({'Promotion': {'No': False, 'Yes': True}}, inplace=True)
df_cleaned['Promotion'] = df_cleaned['Promotion'].astype(bool)
# one-hot encoding
df_position_dummies = pd.get_dummies(df['Product Position'], dtype=bool)
df_cleaned.drop(columns=['Product Position'], inplace=True)
df_cleaned = pd.concat([df_cleaned, df_position_dummies], axis=1)

df_terms_dummies = pd.get_dummies(df['terms'], dtype=bool)
df_cleaned.drop(columns=['terms'], inplace=True)
df_cleaned = pd.concat([df_cleaned, df_terms_dummies], axis=1)

df_cleaned.head()
```

```python
df_cleaned.to_csv("zara_cleaned.csv", index=False)
```
