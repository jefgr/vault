# 2  Example B: Channel Capacity of Independent Parallel Channels
A communication system uses independent parallel channels, each modelled as an AWGN chan-nel.  The bandwidth is identical for all channels.  The signal and noise powers are listed on theadditional page.Answer the following questions.  Explain carefully the assumptions behind your reasoning.

```python
signal_powers = [12, 7, 15, 5, 20, 11]
noise_powers = [5, 18, 9, 30, 4, 14]
total_powers = 70
bandwidth = 25 * 1000 #Hz 25kHz
```

1.  Determine the total information rate of the system for the given signal powers.  Discuss the characteristics of the channels and the assumptions made in your calculation.

```python
from math import log2
Rs = [bandwidth * log2(1 + (power / noise)) for power, noise in zip(signal_powers, noise_powers)]
R_total = sum(Rs)
print(f"R ≤ {R_total /1000 } Kbit/s")
```

2.  You are allowed to redistribute the total signal power across the channels.  Determine a new distribution that increases the total information rate.

```python
print("Iteration 1")
Bx6 = total_powers + sum(noise_powers)
B = Bx6 / 6
print(f"B = {B}")
Px = [B - noise for noise in noise_powers]
print(f"Px = {Px}")
```

```python
Noises_v2 = [noise for noise, Pxi in zip(noise_powers, Px) if Pxi > 0]
print("Iteration 2")
Bx5 = total_powers + sum(Noises_v2)
B_v2 = Bx5 / 5
print(f"B = {B_v2}")
Px_v2 = [B_v2 - noise for noise in Noises_v2]
print(f"Px = {Px_v2}")
```

3.  Compute the resulting information rate under your new power distribution.

```python
Rs = [bandwidth * log2(1 + (power / noise)) for power, noise in zip(Px_v2, Noises_v2)]
R_total = sum(Rs)
print(f"R ≤ {R_total /1000 } Kbit/s")
```

4.  Compare the two information rates and interpret the improvement.

We improved our information rate by giving the noisy channels less energy and the cleaner ones more energy
