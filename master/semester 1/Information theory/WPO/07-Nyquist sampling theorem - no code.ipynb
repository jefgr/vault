# Nyquist sampling theorem


## Entropy of binary encoding: the binary entropy function

Consider $M=4$ periods of a sine and a cosine that has been sampled such that each period contains $N=32$ samples. Plot the signal with its $N$ samples per periode and the start to reduce the number of samples by a factor of two ($2, 4, 8, 16, and 32$ ). Note that above $N/2$ (hence less than 2 sample per period), it is no langer possible to reconstruct the original signal (= the Nyquist sampling theorem).

```python
from math import pi, cos, sin
import matplotlib.pyplot as plt
import numpy as np
```

```python
periods = 4 * 2 * pi
print(f"periods = {periods}")
X = np.linspace(0, round(periods), round(periods) * 20 + 1)
def plot_sin_cos(X):
    Y_cos = [cos(xi) for xi in X]
    Y_sin = [sin(xi) for xi in X]
    return Y_sin, Y_cos

Y_sin, Y_cos = plot_sin_cos(X)

X_32 = np.linspace(0, round(periods), 32 * 4 + 1)
Y_sin_32, Y_cos_32 = plot_sin_cos(X_32)

X_16 = np.linspace(0, round(periods), 16 * 4 + 1)
Y_sin_16, Y_cos_16 = plot_sin_cos(X_16)

X_8 = np.linspace(0, round(periods), 8 * 4 + 1)
Y_sin_8, Y_cos_8 = plot_sin_cos(X_8)

X_4 = np.linspace(0, round(periods), 4 * 4 + 1)
Y_sin_4, Y_cos_4 = plot_sin_cos(X_4)

X_2 = np.linspace(0, round(periods), 2 * 4 + 1)
Y_sin_2, Y_cos_2 = plot_sin_cos(X_2)

X_1 = np.linspace(0, round(periods), 1 * 4 + 1)
Y_sin_1, Y_cos_1 = plot_sin_cos(X_1)

plt.plot(X, Y_cos)
plt.plot(X, Y_sin)
plt.plot(X_32, Y_cos_32)
plt.plot(X_32, Y_sin_32)
plt.scatter(X_32, Y_cos_32)
plt.scatter(X_32, Y_sin_32)
```

```python
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)
plt.plot(X_16, Y_cos_16)
plt.plot(X_16, Y_sin_16)
plt.scatter(X_16, Y_cos_16)
plt.scatter(X_16, Y_sin_16)
```

```python
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)
plt.plot(X_8, Y_cos_8)
plt.plot(X_8, Y_sin_8)
plt.scatter(X_8, Y_cos_8)
plt.scatter(X_8, Y_sin_8)
```

```python
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)
plt.plot(X_4, Y_cos_4)
plt.plot(X_4, Y_sin_4)
plt.scatter(X_4, Y_cos_4)
plt.scatter(X_4, Y_sin_4)
```

```python
plt.plot(X, Y_cos)
plt.plot(X, Y_sin)
plt.plot(X_2, Y_cos_2)
plt.plot(X_2, Y_sin_2)
plt.scatter(X_2, Y_cos_2)
plt.scatter(X_2, Y_sin_2)
```

# Capacity of a PAM modulated signal.

A $1$ Gigabit Ethernet communication uses a PAM-5 modulation with Trellis Code Modulation (TCM) over 4 parallel lines. This implies that it transmits $2$ bits at each sampling instance and this over each line. Note: the 5th level of the PAM-5 is used to obtain a better error correcting capability. What is the minimal signal bandwidth that is required for the transmission of this signal?

```python
C = 1e9 #bit/sample (gigabit)
Camount = 4
TCM = 2 #bits/symbol
fs = C / (Camount * TCM)
fw = fs / 2 # fs  2fw
print(f"fs = {fs / 1e6} MHz")
print(f"fw = {fw / 1e6} MHz")
```
