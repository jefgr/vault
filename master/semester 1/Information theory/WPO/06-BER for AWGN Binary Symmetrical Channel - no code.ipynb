# Bit Error Rate (BER) for Additive White Gaussian Noise (AWGN) Binary Symmetric Channel (BSC)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
```

The BER for for Additive White Gaussian Noise (AWGN) Binary Symmetric Channel (BSC) with signal amplitude $A$ and additive noise variance $\sigma^2$ equals $$ BER =  \frac{1}{2}\mathrm{erfc}\left(\frac{A}{\sqrt{2} \sigma}\right) $$ with the complementary error function $$ \mathrm{erfc}(x)=\frac{2}{\sqrt{\pi}}\int_{x}^{\infty}e^{-t^{2}}dt $$ for $x\ge0 $. 

To link the amplityde and the noise variance to the energy per bit $E_b$ and the noise power spectral density $N_0$, we need to use the relationsship that $$ E_b = A^2 $$ for binary channels and that $$ N_0 = 2\sigma^2 $$ Hence, $$\frac{A^2}{2\sigma^2} = \frac{E_b}{N_0}$$ and therefore
$$ BER =  \frac{1}{2}\mathrm{erfc}\left(\sqrt{\frac{E_b}{N_0}}\right) $$

```python
from math import sqrt
def BER(Eb_N0):
    return 0.5 * sp.erfc(sqrt(Eb_N0))

```

### Note on the conversion to dB scale

A dB scale is a logarithmic scale to represent power (and other) ratio. If $x$ is representing a power ratio $P/P_0$, then the dB scale is $$ x_{dB} = 10 \log_{10}(x_{lin}) $$and $$ x_{lin} = 10^{x_{dB}/10}$$.

If e.g. Voltage ratio's are used, then it is known that $ P \sim V^2 $. Hence, if $x$ is presenting a voltage ratio $V/V_0$, then $$ 20 \log_{10} (x_{lin})$$ needs to be used.

```python
X = np.linspace(0, 12, 1201)
Y = [BER(10 ** (xi / 10)) for xi in X]
print(X)
print(Y)

plt.semilogy(X, Y)
plt.xlabel("Eb/N0 (dB)")
plt.ylabel("BER")
plt.grid(True, which="major")
plt.show()
```
